1. Import Libraries

```python
import pandas as pd
import requests
import os
```

2. Load News Data from Alpha Vantage API

```python
api_key = "XOC14IXT758DB6YX"

url = (
    "https://www.alphavantage.co/query"
    "?function=NEWS_SENTIMENT"
    "&tickers=TSLA"
    "&sort=LATEST"
    "&limit=100"
    f"&apikey={api_key}"
)

response = requests.get(url, timeout=30)

print("Status code:", response.status_code)

data = response.json()

print(data.keys())
```

3. Check API Response

```python
if "feed" in data:
    articles = data["feed"]
    print("Number of articles:", len(articles))
else:
    print("No feed found.")
    print(data)
```

4. Convert API Data to DataFrame

```python
df_news_raw = pd.DataFrame(articles)

print("Raw news shape:", df_news_raw.shape)

display(df_news_raw.head())
```

5. Save Raw News Data

```python
os.makedirs("../data/raw", exist_ok=True)

df_news_raw.to_csv("../data/raw/alpha_vantage_tesla_news_raw.csv", index=False)

print("Raw news data saved to ../data/raw/alpha_vantage_tesla_news_raw.csv")
print("Shape:", df_news_raw.shape)
```

6. Understand the News Data

```python
print("Columns:")
print(df_news_raw.columns)

print("Data types:")
df_news_raw.info()

print("Missing values:")
display(df_news_raw.isnull().sum())
```

7. Select Relevant Columns

```python
columns_to_keep = [
    "title",
    "url",
    "time_published",
    "authors",
    "summary",
    "source",
    "category_within_source",
    "source_domain",
    "topics",
    "overall_sentiment_score",
    "overall_sentiment_label",
    "ticker_sentiment"
]

available_columns = [column for column in columns_to_keep if column in df_news_raw.columns]

df_news_cleaned = df_news_raw[available_columns].copy()

display(df_news_cleaned.head())
```

8. Clean News Data

```python
df_news_cleaned = df_news_cleaned.drop_duplicates(subset=["title", "url", "time_published"])

df_news_cleaned = df_news_cleaned.dropna(subset=["title", "url", "time_published"])

df_news_cleaned["time_published"] = pd.to_datetime(
    df_news_cleaned["time_published"],
    format="%Y%m%dT%H%M%S",
    errors="coerce"
)

df_news_cleaned = df_news_cleaned.dropna(subset=["time_published"])

df_news_cleaned["date"] = df_news_cleaned["time_published"].dt.date

if "overall_sentiment_score" in df_news_cleaned.columns:
    df_news_cleaned["overall_sentiment_score"] = pd.to_numeric(
        df_news_cleaned["overall_sentiment_score"],
        errors="coerce"
    )

df_news_cleaned = df_news_cleaned.sort_values("time_published", ascending=False)

print("Cleaned news shape:", df_news_cleaned.shape)

display(df_news_cleaned.head())
```

9. Save Cleaned News Data

```python
os.makedirs("../data/processed", exist_ok=True)

df_news_cleaned.to_csv("../data/processed/alpha_vantage_tesla_news_cleaned.csv", index=False)

print("Cleaned news data saved to ../data/processed/alpha_vantage_tesla_news_cleaned.csv")
print("Final shape:", df_news_cleaned.shape)
```

10. Simple News Analysis

```python
news_per_day = df_news_cleaned.groupby("date").size().reset_index(name="news_count")

display(news_per_day.head())
display(news_per_day.tail())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(news_per_day["date"], news_per_day["news_count"])
plt.title("Tesla News Count per Day")
plt.xlabel("Date")
plt.ylabel("Number of News Articles")
plt.xticks(rotation=45)
plt.show()
```

```python
if "overall_sentiment_label" in df_news_cleaned.columns:
    sentiment_counts = df_news_cleaned["overall_sentiment_label"].value_counts()

    display(sentiment_counts)

    plt.figure(figsize=(8, 5))
    sentiment_counts.plot(kind="bar")
    plt.title("Tesla News Sentiment Distribution")
    plt.xlabel("Sentiment Label")
    plt.ylabel("Number of Articles")
    plt.xticks(rotation=45)
    plt.show()
```
